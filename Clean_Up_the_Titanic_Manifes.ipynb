Loading CSv file

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
```

Standardize missing tokens

```python
MISSING = {"": pd.NA, " ": pd.NA, "  ": pd.NA, "â€”": pd.NA, "?": pd.NA, "NaN": pd.NA}
df = df.replace(MISSING)
df = df.replace(MISSING)
df
```

Rename columns

```python
df = df.rename(columns={
    "PassengerId": "passenger_id",
    "Pclass": "passenger_class",
    "Gender": "gender",
    "Age": "age",
    "Fare": "fare",
    "Embarked": "embarked"
})
df
```

Format Columns

```python
df['gender'] = df['gender'].str.strip().str.lower().str.capitalize()
df['embarked'] = df['embarked'].str.strip().str.upper()
df['passenger_class'] = (
    df['passenger_class'].str.strip().str.lower()
                .replace({'first': '1', 'second': '2', 'third': '3'})
)
df
```

Data type correction

```python
df["passenger_class"] = pd.to_numeric(df["passenger_class"], errors="coerce")
df["fare"] = pd.to_numeric(
    df["fare"].str.replace("[,$]", "", regex=True),
    errors="coerce"
)
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df
```

remove missing data

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["embarked"] = df["embarked"].fillna('S')
df
```
