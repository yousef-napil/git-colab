```python
import pandas as pd

df = pd.read_csv("daily_steps_30_days.csv")
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))

# Plot the steps for each day (line with dots)
plt.plot(df["day"], df["steps"], marker='o')

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=90)

plt.title("Daily Steps Over 30 Days")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.show()
```
